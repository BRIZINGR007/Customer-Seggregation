### Clustering using K-Means.

- 1. We are considering the age and the spending score and cluster our data.

```python
X1 =  df.loc[:,['Age','Spending Score (1-100)']].values
```

```python
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
    kmeans = KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(X1)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss,linewidth = 2,color="red",marker = "8")
plt.xlabel("K_value")
plt.ylabel("WCSS")
plt.show()
```

```python
# We choose NO. OF CLUSTERS = 4  since there is a elbow band at 4
kmeans = KMeans(n_clusters=4,init="k-means++")

label = kmeans.fit_predict(X1)
label.shape
```

```python
# centroids of the 4 clusters xcordinates,ycordinates
kmeans.cluster_centers_
```

```python
plt.scatter(X1[:,0],X1[:,1], c = kmeans.labels_, cmap="rainbow")
plt.grid()
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color="black")
plt.title("Cluster of Customers")
plt.xlabel("Age")
plt.ylabel('Spending Score (1-100)')
```

- Lets form the cluster between the annual income and the spending score

```python
X2 =  df.loc[:,['Annual Income (k$)','Spending Score (1-100)']].values
```

```python
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
    kmeans = KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(X2)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss,linewidth = 2,color="violet",marker = "8")
plt.xlabel("K_value")
plt.ylabel("WCSS")
plt.show()
```

```python
# We choose NO. OF CLUSTERS = 5  since there is a elbow band near around 5
kmeans = KMeans(n_clusters=5,init="k-means++")

label2 = kmeans.fit_predict(X2)
label2.shape
```

```python
# centroids of the 5 clusters xcordinates,ycordinates
kmeans.cluster_centers_
```

```python
plt.scatter(X2[:,0],X2[:,1],c = label2, cmap="rainbow")
plt.grid()
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color="black")
plt.title("Cluster of Customers")
plt.xlabel("Annual Income(k$)")
plt.ylabel("Spendin Score(1-100)")
```

## Conclusion from the above graph

- Lets form the cluster between the  **'Age'**, 'Annual Income (k$)', 'Spending Score (1-100)'

```python
X3 =  df.iloc[:,1:].values
X3.shape
```

- [:,1:] means we have included all columns and iloc is integer based unlike loc where u have to give the row names

```python
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
    kmeans = KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(X3)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss,linewidth = 2,color="violet",marker = "8")
plt.xlabel("K_value")
plt.ylabel("WCSS")
plt.show()
```

```python
# We choose NO. OF CLUSTERS = 5  since there is a elbow band near around 5
kmeans = KMeans(n_clusters=5,init="k-means++")

label3 = kmeans.fit_predict(X3)
print(label3)
```

```python
print(kmeans.cluster_centers_)
```
